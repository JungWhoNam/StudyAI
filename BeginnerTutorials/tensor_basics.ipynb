## tensor in tensorflow
* 1D: scalar
* 2D: vector
* 3D: matrix

```python
import tensorflow as tf
```

```python
# Initialization of Tensors
tf.constant(4)
tf.constant([1,2,3])
tf.constant([[1,2,3], [4,5,6]])
tf.ones((3,3))
tf.zeros((2,3))
tf.eye(3) # the identity matrix (eye)
tf.random.normal((3,3), mean=0, stddev=1)
tf.random.uniform((3,3), minval=0, maxval=1)
tf.range(start=1, limit=10, delta=2)
```

```python
# type casting
x = tf.range(start=1, limit=10, delta=2)
print(x)
x = tf.cast(x, dtype=tf.float64)
print(x)
```

```python
# Mathematical Operations
x = tf.constant([1,2,3])
y = tf.constant([9,8,7])

z = tf.add(x,y) # or z = x + y
print(z)

z = tf.subtract(x,y) # or z = x - y
print(z)

# element-wise
z = tf.divide(x,y) # or z = x/y
print(z)
z = tf.multiply(x,y) # or z = x*y
print(z)
z = x ** 2
print(z)

# dot-product (element-wise multiplication and summation)
z = tf.tensordot(x, y, axes=1)
print(z)
z = tf.reduce_sum(x*y, axis=0)
print(z)

# matrix multiplication
x = tf.random.normal((2,3))
y = tf.random.normal((3,4))
z = tf.matmul(x,y)
print(z)
z = x @ y
print(z)
```

```python
# Indexing (vector)
x = tf.range(10)
print(x)
print(x[:])
print(x[1:])
print(x[1:3])
print(x[::2])
print(x[::-1])

indices = tf.constant([0,3])
x_ind = tf.gather(x, indices)
print(x_ind)
```

```python
# Indexing (matrix)
x = tf.constant([[1,2],[3,4],[5,6]])
print(x[0])
print(x[0,])
print(x[0:2,:])
```

```python
# Reshaping
x = tf.range(9)
print(x)
x = tf.reshape(x, (3,3))
print(x)

x = tf.transpose(x, perm=[1,0])
print(x)
```
