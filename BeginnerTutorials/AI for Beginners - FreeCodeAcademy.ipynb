# ML Fundamentals

## AI history
* can computer think?
* the first approach: pre-defined rules
* ML is a subset of AI.
    * ML figures out the rules for us.
    * Rather then giving computers a set of rules, the computer figure out the rules from looking into data. 
* NN is a subset of ML and AI.
    * It is a form of ML that uses a layered representation of data.

## Correcting Misconceptions
* AI can be simple or complex; a simple program that competes you in TicTacToe is considered as an AI; it is simulating intellectual human behaviors.
* NN is not modeled after a brain (we don't know much about how the brain works to begin with).

## ML Types
1. Supervised Learning
    * we have both features and labels.
2. Unsupervised Learning
    * we only have features.
    * e.g., clustering data points
3. Reinforcement Learning
    * don't have the data... you have agents, envs, rewards.
    * e.g., agenets beating the super mario game
    * the programmer's job is to figure out the right rewards for the agents working in the envs. The agents is moving towards maximizing the rewards.
