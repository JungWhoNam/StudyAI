# 강의 목표
* constant, variable 이 두개의 tensors의 필요성
* dataset과 같이 immutable 변하지 않는 값을 받는 노드는 constant tensors로 정의
* model parameters와 같이 mutable 변하는 값을 받는 노드는 variable tensors로 정의

```python
import tensorflow as tf
import numpy as np

print(tf.__version__)
```

```python
t1 = tf.Variable([1,2,3])
t2 = tf.constant([1,2,3])

print(t1)
print(t2)
```

```python
print('=====')
print(type(t1))
print(type(t2))
```

constant tensor는 EagerTensor type으로 

variable tensor는 ResourceVariable으로 나옴

```python
# python list, numpy를 tensor로 변환
test_list = [1,2,3]
test_np = np.array([1,2,3])

t1 = tf.constant(test_list)
t2 = tf.constant(test_np)

print('=====')
print(t1)
print(t2)
print(type(t1))
print(type(t2))

t1 = tf.Variable(test_list)
t2 = tf.Variable(test_np)

print('=====')
print(t1)
print(t2)
print(type(t1))
print(type(t2))
```

```python
test_list1 = [10,20,30]
test_list2 = [1,2,3]

t1 = tf.constant(test_list1)
t2 = tf.constant(test_list2)

t3 = t1 + t2
print(type(t3))
```

다른 종류의 tensor 연산 후 결과 값의 type? EagerTensor
