```python
# 1. 파일로부터 데이터 읽어오기

import os
import pandas as pd

# csvFilePath = str(os.path.abspath('')) + '\\csv\\lemonade.csv'
csvFilePath = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
lemonadeData = pd.read_csv(csvFilePath)

# csvFilePath = str(os.path.abspath('')) + '\\csv\\boston.csv'
csvFilePath = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
bostonData = pd.read_csv(csvFilePath)

# csvFilePath = str(os.path.abspath('')) + '\\csv\\iris.csv'
csvFilePath = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
irisData = pd.read_csv(csvFilePath)
```

```python
# 2. 데이터 모양 확인
print(lemonadeData.shape)
print(bostonData.shape)
print(irisData.shape)
```

```python
# 3. 데이터 칼럼이름 확인
print(lemonadeData.columns)
print(bostonData.columns)
print(irisData.columns)
```

```python
# 4. 독립변수와 종속변수 분리
indepVar = lemonadeData[['온도']]
depVar = lemonadeData[['판매량']]
print(indepVar.head())
print(depVar.head())
print('\n')

indepVar = bostonData.copy().drop('medv', axis=1)
depVar = bostonData[['medv']]
print(indepVar.head())
print(depVar.head())
print('\n')

indepVar = irisData.copy().drop('품종', axis=1)
depVar = irisData[['품종']]
print(indepVar.head())
print(depVar.head())
print('\n')
```
